read_holding_registers(reg_addr, reg_nb=1)
Modbus function READ_HOLDING_REGISTERS (0x03)
Parameters
• reg_addr (int) – register address (0 to 65535)
• reg_nb (int) – number of registers to read (1 to 125)

```
from pyModbusTCP.client import ModbusClient
import time
import pandas as pd
import struct

#Modbus TCP
SERVER_HOST = "192.168.1.88"
SERVER_PORT = 502
SERVER_U_ID = 109

c = ModbusClient()

c.host(SERVER_HOST)
c.port(SERVER_PORT)
c.unit_id(SERVER_U_ID)

mylist = []

while True:
    # open or reconnect TCP to server
    if not c.is_open():
        if not c.open():
            print("unable to connect to "+SERVER_HOST+":"+str(SERVER_PORT))

    # if open() is ok, read coils (modbus function 0x01)
    if c.is_open():
       #Tool Counter
        D30 = c.read_holding_registers(4101,2)
        D31 = c.read_holding_registers(4141,2)
        
        Temp1 = D31
        str_temp1 = str(Temp1)
        str_temp1 = str_temp1.replace('[','')
        str_temp1 = str_temp1.replace(']','')
        str_temp1 = str_temp1.replace(',','')
        str_temp2 = str_temp1.split()
        a = int(str_temp2[0])
        b= int(str_temp2[1])

        a= hex(a)
        A = str(a)
        newstrA = A.replace('0x', '')
        newstrA = newstrA.upper()
        temp = 4 - len(newstrA)
        if len(newstrA) < 4: 
          for i in range(temp):
            newstrA = '0'+newstrA
        b = hex(b)
        B = str(b)
        newstrB = B.replace('0x', '')
        newstrB = newstrB.upper()
        temp = 4 - len(newstrB)
        
        if len(newstrB) < 4: 
          for i in range(temp):
            newstrB = '0'+newstrB
        
        C = newstrB+newstrA
        newstrC = struct.unpack('!f', bytes.fromhex(C))[0]
        TempC = newstrC - 273.15
        TempC_2_decimal = "{:.2f}".format(TempC)
        
        mylist.append(TempC_2_decimal)
        
        print('Temp: '+ str(TempC_2_decimal))


    time.sleep(2)
```
